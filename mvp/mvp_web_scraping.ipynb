```python
import pandas as pd
import os
import shutil
```

```python
years = list(range(1991,2023))
```

```python
url_start = "https://www.basketball-reference.com/awards/awards_{}.html"
```

```python
# added time library to prevent being blocked from webscraping
import time
import numpy as np
```

```python
import requests

for year in years:
    url = url_start.format(year)
    data = requests.get(url)
    
    with open("mvp/{}.html".format(year), "w+", encoding="utf-8") as f:
        f.write(data.text)
    lag = np.random.uniform(low=5,high=30)
    time.sleep(lag)
```

```python
from bs4 import BeautifulSoup
```

```python
with open("mvp/1991.html", encoding="utf-8") as f:
    page = f.read()
```

```python
# initialize BeautifulSoup class and pass page string // parser class
soup = BeautifulSoup(page, "html.parser")
```

```python
# remove header row
soup.find('tr', class_="over_header").decompose()
```

```python
# remove everything else by finding specific table
mvp_table = soup.find(id="mvp")
```

```python
# read into pandas
mvp_1991 = pd.read_html(str(mvp_table))[0]
```

```python
mvp_1991
```

```python
dfs = []
for year in years:
    with open("mvp/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_="over_header").decompose()
    mvp_table = soup.find(id="mvp")
    mvp = pd.read_html(str(mvp_table))[0]
    mvp["Year"] = year #create column to differentiate years/pages
    
    dfs.append(mvp)
```

```python
mvps = pd.concat(dfs)
```

```python
mvps.head()
```

```python
# store df in csv format
mvps.to_csv("mvps.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.google.com")
```

```python
import time

player_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html"
year = 1991
url = player_stats_url.format(year)

driver.get(url)
driver.execute_script("window.scrollTo(1,10000)")
time.sleep(2)

html = driver.page_source
```

```python
with open("player/{}.html".format(year), "w+", encoding="utf-8") as f:
    f.write(html)
```

```python
for year in years:
    url = player_stats_url.format(year)

    driver.get(url)
    driver.execute_script("window.scrollTo(1,10000)")
    time.sleep(2)

    html = driver.page_source
    with open("player/{}.html".format(year), "w+", encoding="utf-8") as f:
        f.write(html)
    lag = np.random.uniform(low=5,high=30)
    time.sleep(lag)
```

```python
# save table while getting rid of all header rows
dfs = []
for year in years:
    with open("player/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
        
    soup = BeautifulSoup(page, "html.parser")
    tHeads = soup.find_all('tr', class_="thead")
    for thead in tHeads:
        thead.decompose()
    player_table = soup.find(id="per_game_stats")
    player = pd.read_html(str(player_table))[0]
    player["Year"] = year
    dfs.append(player)
```

```python
players = pd.concat(dfs)
```

```python
players
```

```python
players.to_csv("players.csv")
```

```python
team_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
```

```python
for year in years:
    url = team_stats_url.format(year)

    data = requests.get(url)

    with open("team/{}.html".format(year), "w+", encoding="utf-8") as f:
        f.write(data.text)
    lag = np.random.uniform(low=5,high=30)
    time.sleep(lag)
```

```python
dfs = []
for year in years:
    with open("team/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()

    soup = BeautifulSoup(page, "html.parser")
    tHeads = soup.find_all('tr', class_="thead")
    for thead in tHeads:
        thead.decompose()
    team_table = soup.find(id="divs_standings_E")
    team = pd.read_html(str(team_table))[0]
    team["Year"] = year
    team["Team"] = team["Eastern Conference"]
    del team["Eastern Conference"]
    dfs.append(team)

    soup = BeautifulSoup(page, "html.parser")
    tHeads = soup.find_all('tr', class_="thead")
    for thead in tHeads:
        thead.decompose()
    team_table = soup.find(id="divs_standings_W")
    team = pd.read_html(str(team_table))[0]
    team["Year"] = year
    team["Team"] = team["Western Conference"]
    del team["Western Conference"]
    dfs.append(team)
```

```python
teams = pd.concat(dfs)
```

```python
teams
```

```python
teams.to_csv("teams.csv")
```
